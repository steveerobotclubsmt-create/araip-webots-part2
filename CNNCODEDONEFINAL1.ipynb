```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, callbacks
import matplotlib.pyplot as plt

# 1. DATA PREPARATION (From .ipynb)
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# 2. UPDATED ARCHITECTURE (The "Cat-is-not-a-Plane" Edition)
model = models.Sequential([
    layers.RandomFlip("horizontal", input_shape=(32, 32, 3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.2),

    layers.Conv2D(64, (3, 3), padding='same', activation='elu'),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding='same', activation='elu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.2),

    layers.Conv2D(128, (3, 3), padding='same', activation='elu'),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding='same', activation='elu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.3),

    layers.Conv2D(256, (3, 3), padding='same', activation='elu'),
    layers.BatchNormalization(),
    layers.Conv2D(256, (3, 3), padding='same', activation='elu'),
    layers.GlobalAveragePooling2D(),

    layers.Dense(256, activation='elu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(10)
])

model.summary()

# 3. COMPILE AND TRAIN
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Early stopping to find the sweet spot
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)

history = model.fit(train_images, train_labels, epochs=50, batch_size=64,
                    validation_data=(test_images, test_labels),
                    callbacks=[early_stop])

# --- SAVE THE MODEL ---
# This creates the .keras file that Part 2 will call
model.save('final_model_v3.keras')
print("Model saved as final_model_v3.keras")

# --- PLOT TRAINING HISTORY ---
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.title("Training Progress")
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
from PIL import Image

# Configuration
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

def load_and_prep_image(filename):
    """Loads, crops to square, and resizes for CIFAR-10"""
    img = Image.open(filename)

    # Center Crop to square
    width, height = img.size
    new_size = min(width, height)
    left = (width - new_size)/2
    top = (height - new_size)/2
    img = img.crop((left, top, left + new_size, top + new_size))

    # Resize and Normalize
    img = img.resize((32, 32))
    img_arr = np.array(img) / 255.0

    # Remove Alpha channel if present
    if img_arr.shape[-1] == 4:
        img_arr = img_arr[:,:,:3]

    return img_arr.reshape(1, 32, 32, 3), img

def run_predictions():
    # 1. Load the model saved in Part 1
    model_path = 'final_model_v3.keras'

    if not os.path.exists(model_path):
        print(f"Error: {model_path} not found. Run the training script first!")
        return

    model = tf.keras.models.load_model(model_path)
    print("Model loaded successfully.")

    # 2. List of images to test
    sample_files = ['cat_capture_0.png', 'airplane.png', 'truck.png', 'bird.png']

    for filename in sample_files:
        if os.path.exists(filename):
            img_tensor, img_display = load_and_prep_image(filename)

            # Predict
            pred_logits = model.predict(img_tensor, verbose=0)
            result_idx = np.argmax(pred_logits[0])

            # Apply softmax to get percentage
            probabilities = tf.nn.softmax(pred_logits[0]).numpy()
            confidence = probabilities[result_idx] * 100

            print(f"File: {filename} --> Result: {class_names[result_idx].upper()} ({confidence:.2f}%)")

            # Show Result
            plt.figure(figsize=(2,2))
            plt.imshow(img_display)
            plt.title(f"{class_names[result_idx]} {confidence:.1f}%")
            plt.axis('off')
            plt.show()
        else:
            print(f"Skipping {filename}: File not found.")

if __name__ == "__main__":
    run_predictions()
```
